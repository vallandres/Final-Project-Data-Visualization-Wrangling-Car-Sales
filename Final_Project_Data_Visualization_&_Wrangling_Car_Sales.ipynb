#**Final Project Data Visualization & Wrangling**  

##**Background of Problem**

Perusahaan mobil mengalami peningkatan penjualan yang signifikan dalam beberapa bulan terakhir. Mereka ingin memanfaatkan momentum ini untuk mengoptimalkan kinerja penjualan dan meningkatkan penjualan mereka.
Sehingga tujuannya adalah memanfaatkan peningkatan penjualan untuk mencapai pertumbuhan bisnis sebesar 30% selama 6 bulan kedepan dan memperkuat posisi dealer dalam pasar otomotif.

## **Objective**

Tujuan dari analisis data ini adalah untuk memberikan wawasan yang mendalam tentang kinerja penjualan mobil perusahaan dari tahun 2022 hingga 2023. Fokusnya adalah pada pemahaman tren penjualan, preferensi pelanggan, dan kinerja dealer. Dengan memahami lebih baik data penjualan mereka, perusahaan berharap dapat mengambil langkah-langkah strategis untuk meningkatkan kinerja penjualan dan memperkuat posisi mereka dalam pasar otomotif.

##**Goals**

Berdasarkan tujuan dan latar belakang masalah di atas, beberapa goals yang mungkin dapat diidentifikasi adalah:

1. Mengidentifikasi tren penjualan mobil yang dapat dimanfaatkan untuk meningkatkan strategi pemasaran dan penjualan.
2. Mengidentifikasi preferensi pelanggan.
3. Meningkatkan penjualan mobil sebesar 30% pada tahun 2024 mendatang.

##**Dataset**

Atribut-atribut dalam dataset ini adalah:

1. Car_id : daftar nomor ID mobil
2. Date : tanggal transaksi penjualan mobil.
3. Customer Name : Nama pelanggan yang membeli mobil.
4. Gender : jenis kelamin pelanggan (Laki-laki atau Perempuan).
5. Annual Income : Pendapatan pelanggan.
6. Dealer_Name : Nama dealer mobil.
7. Company : Nama perusahaan atau merek mobil.
8. Model : Nama model mobil.
9. Engine : Spesifikasi mesin mobil.
10. Transmission : Jenis transmisi pada mobil (Otomatis atau Manual).
11. Color : Wwarna exterior mobil.
12. Price ($) : Harga mobil yang dijual.
13. Dealer_No : Nomor dealer yang terkait dengan penjualan.
14. Body Style : Style atau desain bodi mobil (misalnya Sedan, SUV).
15. Phone : Nomor telepon yang terkait dengan penjual mobil.
16. Dealer_Region : Wilayah geografis atau lokasi dealer mobil.

```python
# Import library yang dibutuhkan

# Library pengolahan data
import pandas as pd
import numpy as np

# Library visualisasi
import matplotlib.pyplot as plt
import seaborn as sns

# Ignore warnings untuk mengabaikan atau
# menyembunyikan pesan peringatan (warnings) yang mungkin muncul selama eksekusi program
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read data dari file
df = pd.read_csv ('Car Sales.xlsx - car_data.csv')

# Tampilkan 5 data teratas
df.head()
```

```python
# Dimensi dataset
row, columns = df.shape

print(f"Dataset terdiri atas {row} baris dan {columns} kolom")
```

```python
# Info dataset
df.info()
```

```python
# Menampilkan ringkasan statistik deskriptif dari DataFrame ('df')
df.describe()
```

```python
# Menampilkan angka dalam format desimal standar agar dapat dilihat dengan mudah
pd.options.display.float_format = '{:.2f}'.format

# Menjalankan kembali describe() untuk melihat output dengan format desimal standar
print(df.describe())
```

```python
# Mengecek data duplikat
df.duplicated(keep=False)
```

```python
# Menghitung jumlah duplikat dan non-duplikat
count_duplicates = df.duplicated(keep=False).value_counts()

# Menampilkan hasil
print("Jumlah duplikat:", count_duplicates[True] if True in count_duplicates else 0)
print("Jumlah non-duplikat:", count_duplicates[False] if False in count_duplicates else 0)
```

```python
# Cek frekuensi nan tiap kolom
df.isna().sum()
```

```python
# Cek data pada kolom Date
df.Date.head()
```

```python
# Convert kolom host_since
df.Date = pd.to_datetime(df.Date, yearfirst=True)

# Tampilkan data tanggal setelah diconvert
df.Date
```

```python
# Cek outlier pada kolom Annual Income
# Pilih kolom-kolom yang mengandung outlier
outlier_columns = ['Annual Income']

# Buat boxplot untuk setiap kolom
plt.figure(figsize=(6, 4))
sns.boxplot(data=df[outlier_columns])
plt.title('Boxplot of Outlier Columns')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()
```

```python
# Memperlihatkan oulier dengan bentuk histogram
# Memberikan variasi gambaran untuk melihat distribusi data dan bentuknya
# Buat figure & axes
fig, ax = plt.subplots(figsize=(10, 6))

# Buat histogram plot Annual Income
sns.histplot(data=df,
             x="Annual Income",
             bins=100,
             ax=ax)

plt.show()
```

```python
# Deskripsi statistik dari kolom Annual Income
df["Annual Income"].describe()
```

```python
# Menentukan jika suatu data adalah outlier, apabila nilainya lebih besar dari Q3 + 1.5 IQR
# Cari Q1 & Q3
Q1 = df['Annual Income'].quantile(0.25)
Q3 = df['Annual Income'].quantile(0.75)

print(f"Q1 : {Q1:.2f}")
print(f"Q3 : {Q3:.2f}")
```

```python
# Menghitung IQR
IQR = Q3 - Q1

# Menghitung batas atas (Maximum Boundary)
max_bound = Q3 + 1.5 * IQR

print(f"IQR : {IQR:.2f}")
print(f"Maximum Boundary : {max_bound:.2f}")
```

```python
# Filter data tanpa outlier
df = df[df["Annual Income"] < max_bound]

# Validasi hasil filter
df["Annual Income"].describe()
```

```python
# Cek outlier pada kolom Annual Income setelah difilter
# Pilih kolom yang mengandung outlier
outlier_columns = ['Annual Income']

# Buat boxplot untuk setiap kolom
plt.figure(figsize=(6, 4))
sns.boxplot(data=df[outlier_columns])
plt.title('Boxplot of Outlier Columns')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()
```

```python
# Cek outlier pada kolom Annual Income setelah difilter dengan histogram
# Buat figure & axes
fig, ax = plt.subplots(figsize = (10, 6))

# Buat histogram plot Annual Income
sns.histplot(data = df,
             x = "Annual Income",
             bins = 100,
             ax = ax)

plt.show()
```

```python
# Cek outlier pada kolom Price ($)
# Pilih kolom yang mengandung outlier
outlier_columns = ['Price ($)']

# Buat boxplot untuk kolom
plt.figure(figsize=(6, 4))
sns.boxplot(data=df[outlier_columns])
plt.title('Boxplot of Outlier Columns')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()
```

```python
# Memperlihatkan oulier dengan bentuk histogram
# Memberikan variasi gambaran untuk melihat distribusi data dan bentuknya
# Buat figure & axes
fig, ax = plt.subplots(figsize=(10, 6))

# Buat histogram plot Price
sns.histplot(data=df,
             x="Price ($)",
             bins=100,
             ax=ax)

plt.show()
```

```python
# Deskripsi statistik dari kolom price
df["Price ($)"].describe()
```

```python
# Menentukan jika suatu data adalah outlier, apabila nilainya lebih besar dari Q3 + 1.5 IQR
# Cari Q1 & Q3
Q1 = df['Price ($)'].quantile(0.25)
Q3 = df['Price ($)'].quantile(0.75)

print(f"Q1 : {Q1:.2f}")
print(f"Q3 : {Q3:.2f}")
```

```python
# Menghitung IQR
IQR = Q3 - Q1

# Menghitung batas atas (Maximum Boundary)
max_bound = Q3 + 1.5 * IQR

print(f"IQR : {IQR:.2f}")
print(f"Maximum Boundary : {max_bound:.2f}")
```

```python
# Filter data tanpa outlier
df = df[df["Price ($)"] < max_bound]

# Validasi hasil filter
df["Price ($)"].describe()
```

```python
# Cek outlier pada kolom Annual Income setelah difilter
# Pilih kolom-kolom yang mengandung outlier
outlier_columns = ['Price ($)']

# Buat boxplot untuk setiap kolom
plt.figure(figsize=(6, 4))
sns.boxplot(data=df[outlier_columns])
plt.title('Boxplot of Outlier Columns')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()
```

```python
# Cek outlier pada kolom Price ($) setelah difilter dengan histogram
# Buat figure & axes
fig, ax = plt.subplots(figsize = (10, 6))

# Buat histogram plot price
sns.histplot(data = df,
             x = "Price ($)",
             bins = 100,
             ax = ax)

plt.show()
```
